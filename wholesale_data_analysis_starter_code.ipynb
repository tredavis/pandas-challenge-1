## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def shipping_price(row):
    weight_qty = row['unit_weight'] * row['qty']
    if row['unit_weight'] > 50:
        return 7.00 * weight_qty
    else:
        return 10.00 * weight_qty
    
df['shipping_price'] = df.apply(shipping_price, axis=1)
```

```python
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
total_with_ship = df['subtotal'] + df['shipping_price']
df['total_price'] = total_with_ship + (total_with_ship * 0.0925)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['cost_of_each_line'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df['total_price'] - df['cost_of_each_line']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in clients_top:
    total_spent = df.loc[df['client_id'] == client]['total_price'].sum()
    print(f'Client {client} spent ${total_spent}')
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
df_summary = pd.DataFrame(columns=['client_id', 'qty', 'total_shipping_price', 'total_line_price', 'total_revenue', 'total_profit'])
df_summary['client_id'] = clients_top
df_summary['qty'] = df.groupby('client_id')['qty'].sum().loc[clients_top].values
df_summary['total_shipping_price'] = df.groupby('client_id')['shipping_price'].sum().loc[clients_top].values
df_summary['total_line_price'] = df.groupby('client_id')['cost_of_each_line'].sum().loc[clients_top].values
df_summary['total_revenue'] = df.groupby('client_id')['total_price'].sum().loc[clients_top].values
df_summary['total_profit'] = df.groupby('client_id')['profit'].sum().loc[clients_top].values
df_summary
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns = ['Client ID', 'Units', 'Shipping (millions)', 'Total Revenue (millions)', 'Total Cost (millions)', 'Total Profit (millions)']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    if type(x) == int or type(x) == float:
        return x / 1000000
    else:
        print(f'Error: {x} is not a number.')
        return x



# Apply the currency_format_millions function to only the money columns.
df_summary['total_shipping_price'] = df_summary['total_shipping_price'].map(currency_format_millions)
df_summary['total_line_price'] = df_summary['total_line_price'].map(currency_format_millions)
df_summary['total_revenue'] = df_summary['total_revenue'].map(currency_format_millions)
df_summary['total_profit'] = df_summary['total_profit'].map(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
df_summary.columns = money_columns
df_summary

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
df_summary.sort_values('Total Profit (millions)', ascending=False, inplace=True)
df_summary
```
